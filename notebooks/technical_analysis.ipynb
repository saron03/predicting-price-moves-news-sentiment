```python
import pandas as pd
import os

folder = '../data/yfinance_data'
all_dfs = {}

for file in os.listdir(folder):
    if file.endswith('.csv'):
        symbol = file.split('_')[0]  # e.g., 'AAPL'
        df = pd.read_csv(os.path.join(folder, file), parse_dates=['Date'])
        df = df.sort_values('Date')
        df.set_index('Date', inplace=True)
        all_dfs[symbol] = df

# Example: display AAPL
all_dfs['AAPL'].head()
```

```python
from finta import TA

df = all_dfs['AAPL'].copy()

# Make sure column names match expected format
df.rename(columns={
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'Volume': 'volume'
}, inplace=True)

# Add indicators
df['SMA_20'] = TA.SMA(df, 20)
df['RSI_14'] = TA.RSI(df)
df['MACD'] = TA.MACD(df)['MACD']
df['MACD_signal'] = TA.MACD(df)['SIGNAL']

df.tail()
```

```python
processed_data = {}

for symbol, df in all_dfs.items():
    df = df.copy()
    df.rename(columns={
        'Open': 'open',
        'High': 'high',
        'Low': 'low',
        'Close': 'close',
        'Volume': 'volume'
    }, inplace=True)
    df['SMA_20'] = TA.SMA(df, 20)
    df['RSI_14'] = TA.RSI(df)
    df['MACD'] = TA.MACD(df)['MACD']
    df['MACD_signal'] = TA.MACD(df)['SIGNAL']
    processed_data[symbol] = df
```

```python
import matplotlib.pyplot as plt  # <-- Make sure this is included

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['close'], label='Close Price')
plt.plot(df.index, df['SMA_20'], label='SMA 20', linestyle='--')
plt.title('AAPL Close Price & SMA(20)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 4))
plt.plot(df.index, df['RSI_14'], label='RSI (14)', color='purple')
plt.axhline(70, linestyle='--', color='red')
plt.axhline(30, linestyle='--', color='green')
plt.title('AAPL RSI(14)')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 5))
plt.plot(df.index, df['MACD'], label='MACD', color='blue')
plt.plot(df.index, df['MACD_signal'], label='Signal Line', color='orange')
plt.title('AAPL MACD and Signal Line')
plt.xlabel('Date')
plt.ylabel('MACD')
plt.legend()
plt.grid()
plt.show()
```
